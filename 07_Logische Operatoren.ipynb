Tag 7: Express-Versandprüfung mit 'or' und 'not' #Ein Kunde bekommt kostenlosen Express-Versand, wenn er entweder ein Premium-Mitglied (is_premium = True) ist ODER wenn der Bestellwert über 150 EUR liegt.

```python

bestellwert = float(input("Bitte geben Sie Ihr Bestellwert in Euro ein"))
is_premium_text = str(input("Sind Sie ein Premium-Kunde? (ja/nein) "))

if bestellwert > 0:
     print ("Eingabe gültig, die Berechnung läuft")

     #Einen Text in einen Wahrheitswert (Booleans: True/False) umwandeln
     is_premium = is_premium_text.lower() == "ja"

     if is_premium  or bestellwert > 150:
          versandkosten = 0.0 #EUR
          print("Kostenloser Premium-Versand")
     else:
          versandkosten = 7.5 #EUR
          print("Werden Sie Premium-Kunde, um kostenlosen Versand zu erhalten")
    
     #Berechnung
     if bestellwert >= 500:
          print("premium Satatut, sie erhalten 10% Rabatt")
          rabatt = bestellwert * 0.1
          endwert = bestellwert - rabatt + versandkosten
     elif bestellwert < 500 and bestellwert >= 200:
          print("medium Statut, Sie erhalten 5% Rabatt")
          rabatt = bestellwert * 0.05
          endwert = bestellwert - rabatt + versandkosten
     else:
          print("kein Rabatt möglich")
          rabatt = bestellwert * 0.0
          endwert = bestellwert - rabatt + versandkosten

     #Ausgabe
     print(f"--- Ihre Rechnung ---")
     print(f"Gesamtbetrag: {bestellwert:.2f}")
     print(f"Versandkosten: {versandkosten:.2f}")
     print(f"Rabatt: {rabatt/bestellwert * 100:.2f} %")
     print(f"endwert: {endwert:.2f}")
else:
     print("Eingabe nicht gültig, geben Sie eine positive Wert ein")
```
