#Tag 5 Mehrere Bedingungen verknüpfen (elif)

```python
# Tag 5: Gestaffelte Rabatte mit elif

#Business-Szenario:Warenkorb unter 200 EUR -> 0% RabattWarenkorb 
#zwischen 200 EUR und 500 EUR -> 5% RabattWarenkorb 
#ab 500 EUR -> 10% Rabatt


warenkorb_wert = float(input("Bitte gesamten Warenkorb-Wert eingeben (in EUR): "))

# Mehrere Stufen prüfen
if warenkorb_wert >= 500.0:
    rabatt_satz = 0.10  # 10%
    print("Premium-Status: 10% Rabatt gewährt.")
elif warenkorb_wert >= 200.0 and warenkorb_wert < 500.0:
    rabatt_satz = 0.05  # 5%
    print("Basis-Status: 5% Rabatt gewährt.")
else:
    rabatt_satz = 0.0
    print("Kein Rabatt gewährt.")

# Berechnung und Ausgabe
rabatt_betrag = warenkorb_wert * rabatt_satz
endpreis = warenkorb_wert - rabatt_betrag

print(f"--- Abrechnung ---")
print(f"Rabatt-Satz: {rabatt_satz * 100:.0f}%")
print(f"Abgezogener Rabatt: {rabatt_betrag:.2f} EUR")
print(f"Endpreis zu zahlen: {endpreis:.2f} EUR")
```
