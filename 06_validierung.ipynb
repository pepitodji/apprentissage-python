#Verschachtelte Bedingungen (Nested if)


```python

warenkorb_wert = float(input("Bitte geben Sie Ihre Warenkorb-Wert in Euro ein"))

#Datenvalidierung vor der Berechnung
if warenkorb_wert <=0:
     print("Ungültiger Wert. Bitte geben Sie einen positiven Wert ein.")
else:
     print("Eingabe gültig. Berechne Rabatt...")
     if warenkorb_wert < 200:
         print("Null Rabatt Warenkorb")
         rabatt = 0
         endwert = warenkorb_wert
     elif warenkorb_wert >= 200 and warenkorb_wert < 500:
         print ("Sie erhalten 5 % Rabatt")
         rabatt = warenkorb_wert * 0.05
         endwert = warenkorb_wert - rabatt
     else:
          print("Premieum Status: Sie erhalten 10 % Rabatt")
          rabatt = warenkorb_wert * 0.1
          endwert = warenkorb_wert - rabatt

#Ausgabe
print(f" --- Ihre Rechnung ---")
print(f" Ihre Warenkorb-Wert: {warenkorb_wert:.2f} EUR")
print(f" Ihr rabatt: {rabatt/warenkorb_wert*100:.1f} %")
print(f" Ihre Endwert: {endwert:.2f} EUR")

```
