#Textanalyse im E-Commerce (len, in, replace)

```python
# Tag 10: Textanalyse und Bereinigung von Benutzerdaten
passwort_eingabe = "Geheim123"
kommentar_eingabe = "Diese Software ist echt SCHLECHT und dumm!"

#Passwort-Länge prüfen
passwort_laenge = len(passwort_eingabe)
if passwort_laenge < 8:
     print("Sicherheitswarnung: Dein Passwort ist mit {passwort_laenge} zeichen" )
else:
     print("PasswortLänge ist ausreichen")

# 2. Schimpfwörter blockieren oder filtern mit 'in' und '.replace()'
unhoefliches_wort = "SCHLECHT"
if unhoefliches_wort in kommentar_eingabe:
     print("\nWarnung: unhöfliches Wort in Kommentar endeckt!")

     #Das Wort durch Sternschen ersetzen
     sauberer_kommentar = kommentar_eingabe.replace(unhoefliches_wort, "S***")
     print(f"Zensierter Kommentar: {sauberer_kommentar}")

```
