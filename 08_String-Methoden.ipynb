Tag 8: Bereinigung von Kundendaten

```python
eingabe_name = "max MUSTERMANN"
eingabe_email = "max.MUSTERMANN@wirtschaftsinformatik.de"


# Text bereinigen
# .strip() entfernt überflüssige Leerzeichen am Anfang und Ende
# .title() macht den ersten Buchstaben groß, den Rest klein
# .lower() macht alles klein (perfekt für E-Mails)
sauberer_name = eingabe_name.strip().title()
saubere_email = eingabe_email.strip().lower()

#Ausgabe
print("--- Eingabe Datenbank ---")
print(f"Name: {sauberer_name}")
print(f"E-Mail: {saubere_email}")
```
