#Mathematische Operationen im Business-Kontext
#Modulo, Ganzzahl-Division, Potenz

```python
#Tag 3: Business-Mathematik

kapital = 1000.0   # Startkapital in Euro
zinssatz = 0.05   # Zinssatz pro Kalenderjahr in Dezimalform (5% = 0.05) 
laufzeit = 3      # Laufzeit in Jahren

#zinseszins berechnen mit dem Potenz-Operator (**)
endkapital = kapital * (1+zinssatz) ** laufzeit
gewinn = endkapital - kapital

print(f"--- Zinseszins Rechner ---")
print(f"startkapital: {kapital:.2f} EUR--")
print(f"endkapital nach {laufzeit} Jahren: {endkapital:.2f} EUR--")
print(f"Reiner Zinsgewinn: {gewinn:.2f} EUR--")

```
