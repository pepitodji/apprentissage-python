#Logik, Bedingungen if / elif / else


```python
# Tag 4: Bedingungen (if/else) im Business. #Fragestellung: Wenn ein Kunde für mehr als 500 Euro einkauft, 
# erhält er 10 % Rabatt. Wenn nicht, zahlt er den vollen Preis.

warenkorb_wert = float(input("Bitte gesamten Warenkorb-Wert eingeben (in EUR): "))

#Entscheidung treffen
if warenkorb_wert >= 500.0:
     print(f"Grückwunsch! Du erhältst 10% Premieum-Rabatt.")
     rabatt = warenkorb_wert * 0.10
     endpreis = warenkorb_wert - rabatt
else:
     print(f"Kein Rabatt möglich (Mindestbestellwert für Rabatt: 500 EUR).")
     endpreis = warenkorb_wert

#Ausgabe
print(f"---Abrechnung---")
print(f"Abgezogenes Rabatt: {rabatt:.2f} EUR")
print(f"Endpreis zu zahlen: {endpreis:.2f} EUR")
```
