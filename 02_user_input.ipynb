```python
# 1. Daten vom Nutzer abfragen (Werte sind zuerst Text!)
produkt_name = input("Bitte Produktname eingeben: ")
preis_text = input("Bitte Netto-Preis eingeben (z.B. 100.50): ")
menge_text = input("Bitte Menge eingeben (z.B. 5): ")

# 2. Typkonvertierung der Preise und Mengen: Test in Zahlen umwandeln
# float() macht daraus eine Dezimalzahl, int() macht daraus eine ganze Zahl
preis = float(preis_text)
menge = int(menge_text)

# 3. Berechnung des Gesamtpreises
netto_gesamt = preis * menge
mehrwertsteuer = netto_gesamt * 0.19  # 19% Mehrwertsteuer
brutto_gesamt = netto_gesamt + mehrwertsteuer

# 4. Ausgabe der Ergebnisse
print(f"\n--- Rechnung für {produkt_name} ---")
print(f"Gesamt Netto: {netto_gesamt:.2f} EUR")
print(f"Mehrwertsteuer (19%): {mehrwertsteuer:.2f} EUR")
print(f"Gesamt Brutto: {brutto_gesamt:.2f} EUR")
menge = int(menge_text)

```
